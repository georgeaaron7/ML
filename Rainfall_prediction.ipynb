```python
import sklearn as sk
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("austin_weather_final.csv")
data
```

```python
X = data.drop(["PrecipitationSumInches"],axis=1,)
```

```python
Y = data["PrecipitationSumInches"]
```

reshaping it into a 2d vector

```python
Y = Y.values.reshape(-1,1)
```

```python
day_index = 798
days = [i for i in range(Y.size)]
```

intializing the linear regression model

```python
clf = LinearRegression()
```

Training the clf model

```python
clf.fit(X,Y)
```

plotting graph

```python
print("The precipitation trend graph")
plt.scatter(days,Y,color = "r")
plt.scatter(days[day_index],Y[day_index],color="b")
plt.title("Precipitation Level")
plt.xlabel("Days")
plt.ylabel("Precipitation in Inches")
plt.show
x_vis = X.filter(['TempAvgF', 'DewPointAvg', 'HumidityAvgPercent', 'SeaLevelPressureAvgInches', 'VisibilityAvgMiles', 'WindAvgMPH'])
```

```python
print("The precipitation vs attribrutes trend graph")

for i in range(x_vis.columns.size):
  plt.subplot(3,2,i+1)
  plt.scatter(days,x_vis[x_vis.columns.values[i][:100]] , color = "r")
  plt.scatter(days[day_index],x_vis[x_vis.columns.values[i]][day_index], color = "b")
  plt.title(x_vis.columns.values[i])
plt.show()
```
